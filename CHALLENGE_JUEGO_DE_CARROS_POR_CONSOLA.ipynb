```python
import datetime
import time
import random
import json
import os
class Jugador():
  contarNombre=0
  nombre=[]
    
class Conductor():
  contarConduc=Jugador.contarNombre
 
 
class Carril():
  contarCarril=Conductor.contarConduc
 
class Pista():
  _distanciaKm=2
  posicionCarro=0
  tiempoLlegada=[]
 
class Carro():
  contarCarros=Carril.contarCarril
  listaCarro=[]
    
class Juego():
  IDjuego=1
  podioLista=[]
  podiocontador=[]
  def configurar_juego():
    print("vamos a configurar el juego, debes registrar por lo menos 3 jugadores ")
    while True:      
      Jugador.ingresenombre=input("Ingrese su nombre nuevo jugador: ")
      Jugador.otrojuegador=int(input("¿Desea agregar otro jugador? 1(Si) 2(No): "))
      Jugador.contarNombre=Jugador.contarNombre+1
      Conductor.contarConduc=Conductor.contarConduc+1
      Carro.contarCarros=Carro.contarCarros+1
      Carril.contarCarril=Carril.contarCarril+1
      Jugador.nombre.append(Jugador.ingresenombre)
      if Jugador.otrojuegador == 2:
        break
    for i in range(len(Jugador.nombre)):
      Juego.podiocontador.append(0)

  def iniciar_juego():
    
    for Pista.posicionCarro in range(Carril.contarCarril):
      while Pista.posicionCarro < Pista._distanciaKm*1000:
        Pista.posicionCarro=Pista.posicionCarro + random.randint(1,6)*100            
      Carro.listaCarro.append(Pista.posicionCarro)
      Pista.posicionCarro = random.uniform(0,time.time())
      Pista.tiempoLlegada.append(Pista.posicionCarro)

   
  def asignar_podio():
    orden_de_llegada = [x for _,x in sorted(zip(Pista.tiempoLlegada,Jugador.nombre))]
    print("El primer lugar es para: ", orden_de_llegada[0])
    print("El segundo lugar es para: ", orden_de_llegada[1])
    print("El tercer lugar es para: ", orden_de_llegada[2])
    Juego.podioLista.append(orden_de_llegada)
    for i in range(len(Jugador.nombre)): #en base a la lista de nombres
      if Jugador.nombre[i] == orden_de_llegada[0]: #compara con el numero 1 de esa carrera
        Juego.podiocontador[i] += 1
      print(Jugador.nombre[i],"-----", Juego.podiocontador[i])
    


  
  def nuevo_juego():
    while True:      
      jugarDeNuevo=int(input("¿Desea jugar otra partida? 1(Si) 2(No): "))
      Pista.tiempoLlegada=[]
      Juego.IDjuego=Juego.IDjuego+1
      Juego.iniciar_juego()
      Juego.asignar_podio()
      if jugarDeNuevo == 2:
        break

  
 

class Podio(Juego):
  def guardar():
    puntajes = []
    
    if os.path.exists('./leaderboaard.json'):
      with open('puntajes.json', 'r') as file:
        puntajes = json.load(file)
    if len(puntajes) == 0:
      for i in range(len(Jugador.nombre)):
        puntajes.append([Jugador.nombre[i],Juego.podiocontador[i]])

    else:
      for i in range(len(Jugador.nombre)):
        for j in range(len(puntajes)):
          if Jugador.nombre[i] == puntajes[j][0]:
            puntajes[j][1] += Juego.podiocontador[i]
            break
          elif j ==len(puntajes) - 1:
            puntajes.append([Jugador.nombre[i],Juego.podiocontador[i]])
    puntajes = sorted(puntajes, key=lambda score: score[1], reverse=True)

    with open('puntajes.json', 'w') as file:
      json.dump(puntajes, file)

  def tablaglobal():

    with open('puntajes.json', 'r') as file:
      puntajes = json.load(file)
    print("nombres---podios")
    for i in range(len(puntajes)):
      print(puntajes[i][0],"----",puntajes[i][1])
 

#Prints de pruebas, para verificar el correcto funcionamiento de las funciones 
Juego.configurar_juego()
Juego.iniciar_juego()
Juego.asignar_podio()
Juego.nuevo_juego()
#Podio.guardar_datos()
#print("La cantidad total de jugadores es: ",Jugador.contarNombre)
#print("La cantidad total de conductores es: ",Conductor.contarConduc)
#print("La cantidad total de carriles es: ",Carril.contarCarril)
#print("La cantidad total de carros es: ",Carro.contarCarros)
#print(Jugador.nombre)
#print(Carro.listaCarro)
#print(Pista.tiempoLlegada)
print("Se jugaron ",Juego.IDjuego," partidas")
Podio.guardar()
Podio.tablaglobal()
```
